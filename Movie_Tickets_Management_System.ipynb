```python
class Movie:
    def __init__(self, movieId, title, ticket_price, num_of_tickets, language, movie_3D):
        self.movieId = movieId
        self.title = title
        self.ticket_price = ticket_price
        self.num_of_tickets = num_of_tickets
        self.language = language
        self.movie_3D = movie_3D

    def calculate_total_price(self):
        return self.num_of_tickets * self.ticket_price

    def get_3d_charge(self):
      if self.movie_3D:
        return 50
      else:
        return 0

    def get_gst_rate(self):
        if self.language.lower() == 'hindi':
            return 8
        elif self.language.lower() == 'telugu':
            return 5
        elif self.language.lower() == 'english':
            return 10
        elif self.language.lower() == 'tamil':
            return 10
        else:
            return 5  # Default GST rate

    def calculate_gst(self, total_price):
        gst_rate = self.get_gst_rate()
        return total_price * (gst_rate / 100)

    def display_bill(self):
        total_price = self.calculate_total_price()
        gst_amount = self.calculate_gst(total_price)
        three_d_charge = self.get_3d_charge()
        final_bill = total_price + gst_amount + three_d_charge

        print(f"Movie ID: {self.movieId}")
        print(f"Movie Language: {self.language}")
        print(f"3D Movie       : {'Yes' if self.movie_3D else 'No'}")
        print(f"Title: {self.title}")
        print(f"Ticket Price: {self.ticket_price}")
        print(f"Number of Tickets: {self.num_of_tickets}")
        print(f"Total Price: {total_price}")
        if self.movie_3D:
            print(f"3D Charge: {three_d_charge}")
        print(f"GST Amount: {gst_amount}")
        print(f"Final Bill: {final_bill}")

cus1 = Movie(1, "LEO", 250, 2, "telugu", True)
cus1.display_bill()
```
