```python
class FoodItem:
  def food_information(self,item_Id,item_name,price,quality):
    self.item_Id=item_Id
    self.item_name=item_name
    self.price=price
    self.quality=quality

  def calculate_total_bill(self,quantity):
    return self.price * quantity

  def packing_charge(self):
    return 20

  def delivery_charge(self,quantity):
    return quantity * 20

  def calculate_gst(self, total_bill, gst_rate=5):
    return total_bill * (gst_rate / 100)

  def display_bill(self, quantity, discount_percentage=0):
    total_bill = self.calculate_total_bill(quantity)
    discount_amount = total_bill * (discount_percentage / 100)
    gst_amount = self.calculate_gst(total_bill)
    packing_bill = self.packing_charge()
    delivery_bill = self.delivery_charge(quantity)
    final_bill = total_bill - discount_amount + gst_amount + packing_bill + delivery_bill

    print(f"item_Id:{self.item_Id}")
    print(f"item_name:{self.item_name}")
    print(f"quality:{self.quality}")
    print(f"price:{self.price}")
    print(f"quantity:{quantity}")
    print(f"total_bill:{total_bill}")
    print(f"discount_applied: {discount_percentage}%")
    print(f"discount_amount:{discount_amount}")
    print(f"delivery_charges:{delivery_bill}")
    print(f"gst_amount:{gst_amount}")
    print(f"packing_bill:{packing_bill}")
    print(f"final_bill:{final_bill}")

item1 = FoodItem()
item1.food_information(1, "Burger", 500, "Excellent")

item2 = FoodItem()
item2.food_information(2, "pizza", 500, "good")

item1.display_bill(1, 10) # 2 quantities, 10% discount
print("\n")
item2.display_bill(2,) # 2 quantities, no discount
```
