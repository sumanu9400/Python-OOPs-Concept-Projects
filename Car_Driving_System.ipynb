```python
# Base class
class Car:
    def start_engine(self):
        pass

    def shift_gear(self):
        pass

    def acceleration_engine(self):
        pass

    def break_car(self):
        pass

    def drive_car(self):
        self.start_engine()
        self.shift_gear()
        self.acceleration_engine()
        self.break_car()


# Subclass for Tesla
class TeslaElectric(Car):
    def start_engine(self):
        print("Tesla engine started silently")

    def shift_gear(self):
        print("Tesla shift gear")

    def acceleration_engine(self):
        print("Tesla accelerating smoothly")

    def break_car(self):
        print("Tesla regenerative braking activated")


# Subclass for BMW
class BMW(Car):
    def start_engine(self):
        print("BMW engine started with a roar")

    def shift_gear(self):
        print("BMW shift gear")

    def acceleration_engine(self):
        print("BMW accelerating powerfully")

    def break_car(self):
        print("BMW hydraulic brakes applied")


# Usage
tesla = TeslaElectric()
bmw = BMW()

print("Driving Tesla:")
tesla.drive_car()

print("\nDriving BMW:")
bmw.drive_car()
```
